```python

import numpy as np
from qiskit import QuantumCircuit
import  matplotlib.pyplot as plt
import pickle , json
```

```python

def cnot_count(qc: QuantumCircuit):
    count_ops = qc.count_ops()
    if 'cx' in count_ops:
        return count_ops['cx']
    return 0


def cz_count(qc: QuantumCircuit):
    count_ops = qc.count_ops()
    if 'cz' in count_ops:
        return count_ops['cz']
    return 0
```
