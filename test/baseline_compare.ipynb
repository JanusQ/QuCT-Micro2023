```python
from collections import  defaultdict
import numpy as np
def avg(points):
    x2y = defaultdict(list)
    for x, y in points:
        if x == 7:
            continue
        x2y[x].append(y)

    x_ys = []
    for x, ys in x2y.items():
        x_ys.append([x, sum(ys)/len(ys)])

    x_ys.sort(key = lambda key: key[0])

    return np.array([x for x, y in x_ys]), np.array([y for x, y in x_ys])

cpu_x, cpu_y = avg([[result['n_neighbors'], result['cpu time']] for result in results])
time_x, time_y = avg([[result['n_neighbors'], result['synthesis_time']] for result in results])

fig1, ax1 = plt.subplots(figsize =(6,6))
ax1.plot(cpu_x, cpu_y/3600, label = 'cpu time')
# ax1.plot(time_x, time_y, label = 'syn time')
```
