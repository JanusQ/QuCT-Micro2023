```python
import pickle
import numpy as np
y = pickle.load(open('cut_type_y.pkl','rb'))
```

```python
def exponential_func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
x_data = np.arange(1, 9)
y_data =  y[1:]
```

```python
from scipy.optimize import curve_fit
params, params_covariance = curve_fit(exponential_func, x_data, y_data)
a_fit, b_fit, c_fit = params
x_fit = np.arange(9, 60)
y_fit = exponential_func(x_fit, a_fit, b_fit, c_fit)
```

```python
x_fit,y_fit
```

```python
import matplotlib.pyplot as plt 
import numpy as np
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x_fit, y_fit)
# fig.savefig('bar.svg')
```
